# Popuraity Based Recommendation System

```python
import numpy as np
import pandas as pd
```

```python
movie_name = pd.read_csv('movies.csv')
movie_name.head()
```

```python
ratings_data = pd.read_csv('ratings.csv')
ratings_data
```

```python
movie_data = pd.merge(movie_name , ratings_data,on = 'movieId')
movie_data.head()
```

# Criteria For Popularity Based Recommendation System 

The criteria is based on:


Movies with the highest rating.

Number of views

# Calculating average ratings

```python
ratings_mean_count = pd.DataFrame(movie_data.groupby('title')['rating'].mean())
ratings_mean_count
```

```python
ratings_mean_count['rating_counts'] = pd.DataFrame(movie_data.groupby('title')['rating'].count())
```

```python
ratings_mean_count
```

```python
ratings_mean_count['rating'] = round(ratings_mean_count['rating'],1)
```

```python
ratings_mean_count
```

```python
ratings_mean_count
```

```python
ratings_mean_count = ratings_mean_count[(ratings_mean_count['rating'] >3) & (ratings_mean_count['rating_counts'] > 100)]
ratings_mean_count
```

```python
ratings_mean_count = ratings_mean_count.sort_values(by = 'rating' , ascending = False)
ratings_mean_count
```
